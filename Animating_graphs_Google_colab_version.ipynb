## Animating a graph and producing a video out of it

This code is designed to produce an animation of graphs. If we want to see how a graph changes with a parameter in it, we can animate the graph with the changing parameter. In this example below, we animate the graph $y=\sin(2\pi x)+0.1r$ with $r$ changing, which will move the graph vertically. The code saves the output of the animation as an mp4 file, and displays it on the screen in a loop.

These vertical shifts of a given graph can be thought of as various antiderivatives of the function $f(x)=-\frac{1}{2\pi} \cos(2\pi x)$. 

```python
# Modified from https://www.geeksforgeeks.org/using-matplotlib-for-animations/

# Enable interactive plot; allows stopping the animation
from matplotlib import rc
rc('animation', html='jshtml')
from matplotlib import pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d
import matplotlib.animation as animation

fig = plt.figure() #initial empty plot

axis = plt.axes(xlim =(0, 4), ylim =(-1, 4)) #range for x, y values

line, = axis.plot([], [], lw = 3) # initializing a line variable

# data which the line will
# contain (x, y)
def init():
	line.set_data([], [])
	return line,

def animate(r):
	x = np.linspace(0, 4, 1000)
	y = np.sin(2 * np.pi * x)+0.1*r # plots a sine graph y=sin(2pi x) with 0.1r added
	line.set_data(x, y)
	return line,

anim = animation.FuncAnimation(fig, animate, init_func = init,
					frames = 50, interval = 50, blit = True, repeat=True)

#if you don't want to save a copy of the animation, ignore next line
#anim.save('sinewaveupdown.mp4', writer = 'ffmpeg', fps = 20)

anim
```
