## Plotting Functions  

In this section, we learn how to plot functions in Python. We first need to have a basic function to plot. Let's take sin(x). Any fancy function we use (i.e. anything but polynomials), will come from numpy package. So we will refer to them as "np.name" in the code. For example, y=sin(x) function will be np.sin(x). When plotting, we also need to indicate what we want for the x range and that's done in the line using "np.arange".

```python
import matplotlib.pyplot as plt #import packages; first for plotting, second
import numpy as np              #for fancy math functions

x = np.arange(-2*np.pi, 2*(np.pi), 0.1) # range for x values
y = np.sin(x) # assign y-coordinates based on a formula; change function here
  
plt.plot(x, y, "g") # plot the points with color g(reen)
  
plt.show() 
```

Polynomials can be written as y=x\*\*3-4\*x\**2, for y=x^3-4x^2 and does not require a call to numpy. (If running just this portion, add the top matter from the first bit to top of here too.)

```python
x = np.arange(0, 4, 0.1) 
y = x**2+1 
  
plt.plot(x, y, "g") # plotting the points with color g(reen)
  
plt.show()
```

If we want to plot multiple functions on the same set of axes, we use fancier stuff. (If running just this portion, add the top matter from the first bit to top of here too.)

```python
x = np.linspace(0, 3, 10)  # range for x values; using a different style

plt.plot(x, x**2+1, label='f(x)')  # skip defining y, just use formula 
plt.plot(x, np.sin(x), label='g(x)')  # second function
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Two in one")
plt.legend(loc='best') #just to remove a weird legend warning
plt.show()
```
