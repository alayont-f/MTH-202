```python
# Modified from https://www.geeksforgeeks.org/using-matplotlib-for-animations/

# Enable interactive plot; allows stopping the animation
%matplotlib notebook

from matplotlib import pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation

fig = plt.figure() #initial empty plot

axis = plt.axes(xlim =(0, 4), ylim =(-1, 4)) #range for x, y values

line, = axis.plot([], [], lw = 3) # initializing a line variable

# data which the line will
# contain (x, y)
def init():
	line.set_data([], [])
	return line,

def animate(r):
	x = np.linspace(0, 4, 1000)
	y = np.sin(2 * np.pi * x)+0.1*r # plots a sine graph y=sin(2pi x) with 0.1r added
	line.set_data(x, y)
	return line,

anim = FuncAnimation(fig, animate, init_func = init,
					frames = 50, interval = 50, blit = True, repeat=True)

#if you don't want to save a copy of the animation, ignore next line
anim.save('sinewaveupdown.mp4', writer = 'ffmpeg', fps = 20)


plt.show()
```
